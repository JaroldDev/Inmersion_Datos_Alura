
1. Promedio del área de todos los inmuebles en los barrios en el dataset. El top 10;

```python
import pandas as pd
inmuebles = pd.read_csv('inmuebles_bogota.csv')
inmuebles.head()
```

```python
columnas = {'Baños':'Banos','Área':'Area'}
inmuebles = inmuebles.rename(columns=columnas)
inmuebles.sample(10)
```

```python
inmueblesPromBarrio = inmuebles.groupby('Barrio').Area.mean().sort_values(ascending=False)
inmueblesPromBarrio.head(10)
```

```python
inmueblesPromBarrio.head(10).plot.bar(color='red')
```
